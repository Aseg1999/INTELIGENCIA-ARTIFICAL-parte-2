```python
import time
from IPython.display import clear_output

# ======================================
# TEORÍA DE JUEGOS - CÓDIGO 6
# COOPERAR O TRAICIONAR
# ======================================

escenarios = [
    ("cooperar", "cooperar"),
    ("cooperar", "traicionar"),
    ("traicionar", "cooperar"),
    ("traicionar", "traicionar")
]

# --------------------------------------
# DIBUJAR ESCENA
# --------------------------------------
def mostrar_escena(accion_a, accion_b, oro_a, oro_b):
    print("╔══════════════════════════════════════╗")
    print("║       REPARTO DEL BOTÍN FINAL        ║")
    print("╚══════════════════════════════════════╝\n")

    print("⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛")
    print("⬛ 🧙 ⬜ ⬜ ⬜ 👑 ⬜ ⬜ ⬜ 🧝 ⬛")
    print("⬛ ⬜ ⬜ ⬜ 💰 💰 💰 ⬜ ⬜ ⬜ ⬛")
    print("⬛ ⬜ ⬜ ⬜ 💎 🏆 💎 ⬜ ⬜ ⬜ ⬛")
    print("⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛")
    print()

    print(f"🧙 Jugador A | Estrategia: {accion_a} | Oro ganado: {oro_a}")
    print(f"🧝 Jugador B | Estrategia: {accion_b} | Oro ganado: {oro_b}")
    print()

# --------------------------------------
# RECORRER ESCENARIOS
# --------------------------------------
for accion_a, accion_b in escenarios:

    oro_a = 0
    oro_b = 0

    clear_output(wait=True)
    print("TEORÍA DE JUEGOS - COOPERAR O TRAICIONAR\n")
    print("PASO 1: ESCENARIO INICIAL\n")
    mostrar_escena(accion_a, accion_b, oro_a, oro_b)

    print("Ambos jugadores están decidiendo qué hacer con el botín...")
    time.sleep(4)

    # Reglas del juego
    if accion_a == "cooperar" and accion_b == "cooperar":
        oro_a += 50
        oro_b += 50
        resultado = "Ambos cooperaron y compartieron el botín de forma justa."

    elif accion_a == "cooperar" and accion_b == "traicionar":
        oro_a += 10
        oro_b += 90
        resultado = "Jugador B traicionó y se quedó con casi todo el botín."

    elif accion_a == "traicionar" and accion_b == "cooperar":
        oro_a += 90
        oro_b += 10
        resultado = "Jugador A traicionó y obtuvo casi todo el botín."

    else:
        oro_a += 20
        oro_b += 20
        resultado = "Ambos traicionaron. Se rompió la confianza y el botín se perdió en gran parte."

    clear_output(wait=True)
    print("TEORÍA DE JUEGOS - COOPERAR O TRAICIONAR\n")
    print("PASO 2: RESULTADO DEL ESCENARIO\n")
    mostrar_escena(accion_a, accion_b, oro_a, oro_b)

    print("Resultado:")
    print(resultado)

    time.sleep(6)

clear_output(wait=True)
print("TODOS LOS ESCENARIOS FUERON EVALUADOS\n")
print("Conclusión:")
print("Cuando dos jugadores toman decisiones estratégicas, el resultado cambia según lo que haga cada uno.")
print("Eso representa la idea principal de la teoría de juegos.")
```
