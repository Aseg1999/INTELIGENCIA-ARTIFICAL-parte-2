```python
import time
from IPython.display import clear_output

# ======================================
# HEURÍSTICA VISUAL
# ENEMIGO PERSIGUE AL JUGADOR
# ======================================

filas = 7
columnas = 9

enemigo = (1, 1)
jugador = (5, 7)

recorrido = []

# --------------------------------------
# FUNCIÓN HEURÍSTICA
# --------------------------------------
def heuristica(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

# --------------------------------------
# DIBUJAR MAPA
# --------------------------------------
def dibujar_mapa(enemigo, jugador, recorrido):

    for i in range(filas):
        fila_visual = ""

        for j in range(columnas):

            pos = (i, j)

            if pos == enemigo:
                fila_visual += "🤖 "

            elif pos == jugador:
                fila_visual += "😀 "

            elif pos in recorrido:
                fila_visual += "⭐ "

            else:
                fila_visual += "⬜ "

        print(fila_visual)

    print()

# --------------------------------------
# MOVIMIENTOS POSIBLES
# --------------------------------------
movimientos = [
    (0,1),   # derecha
    (1,0),   # abajo
    (0,-1),  # izquierda
    (-1,0)   # arriba
]

# --------------------------------------
# INICIO
# --------------------------------------
print("HEURÍSTICA VISUAL: ENEMIGO PERSIGUE AL JUGADOR\n")
time.sleep(2)

paso = 1

while enemigo != jugador:

    clear_output(wait=True)

    print(f"PASO {paso}\n")
    dibujar_mapa(enemigo, jugador, recorrido)

    print(f"Posición enemigo: {enemigo}")
    print(f"Posición jugador: {jugador}")

    time.sleep(2)

    mejor_posicion = None
    mejor_distancia = float("inf")

    # evaluar movimientos
    for mov in movimientos:

        nueva = (enemigo[0] + mov[0], enemigo[1] + mov[1])

        if 0 <= nueva[0] < filas and 0 <= nueva[1] < columnas:

            distancia = heuristica(nueva, jugador)

            if distancia < mejor_distancia:

                mejor_distancia = distancia
                mejor_posicion = nueva

    # guardar recorrido
    recorrido.append(enemigo)

    # mover enemigo
    enemigo = mejor_posicion

    # mostrar movimiento
    clear_output(wait=True)

    print(f"PASO {paso} - MOVIMIENTO\n")
    dibujar_mapa(enemigo, jugador, recorrido)

    print(f"El enemigo se movió a {enemigo}")
    print(f"Distancia al jugador: {mejor_distancia}")

    paso += 1

    time.sleep(2)

# --------------------------------------
# RESULTADO FINAL
# --------------------------------------
clear_output(wait=True)

print("RESULTADO FINAL\n")

dibujar_mapa(enemigo, jugador, recorrido)

print("El enemigo alcanzó al jugador 😀")
print(f"Posición final: {enemigo}")
```
