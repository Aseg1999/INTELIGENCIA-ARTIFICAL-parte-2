```python
import time
from IPython.display import clear_output

# ======================================
# TEORÍA DE JUEGOS - CÓDIGO 5
# DOS EQUIPOS COMPITEN POR RECURSOS
# ======================================

escenarios = [
    ("recoger", "recoger"),
    ("recoger", "atacar"),
    ("atacar", "recoger"),
    ("atacar", "atacar")
]

# --------------------------------------
# DIBUJAR RECURSOS
# --------------------------------------
def mostrar_mapa():
    print("╔══════════════════════════════════════╗")
    print("║      ZONA DE RECURSOS DEL MAPA       ║")
    print("╚══════════════════════════════════════╝\n")
    print("⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛")
    print("⬛ 🔴 ⬜ ⬜ 💎 💎 💎 ⬜ ⬜ 🔵 ⬛")
    print("⬛ ⬜ ⬜ ⬜ 💎 🏆 💎 ⬜ ⬜ ⬜ ⬛")
    print("⬛ ⬜ ⬜ ⬜ 💎 💎 💎 ⬜ ⬜ ⬜ ⬛")
    print("⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛ ⬛")
    print()

# --------------------------------------
# MOSTRAR ESTADO
# --------------------------------------
def mostrar_estado(accion_rojo, accion_azul, puntos_rojo, puntos_azul):
    mostrar_mapa()

    print(f"🔴 Equipo Rojo   | Estrategia: {accion_rojo} | Puntos: {puntos_rojo}")
    print(f"🔵 Equipo Azul   | Estrategia: {accion_azul} | Puntos: {puntos_azul}")
    print()

# --------------------------------------
# RECORRER ESCENARIOS
# --------------------------------------
for accion_rojo, accion_azul in escenarios:

    puntos_rojo = 0
    puntos_azul = 0

    clear_output(wait=True)
    print("TEORÍA DE JUEGOS - COMPETENCIA POR RECURSOS\n")
    print("PASO 1: ESCENARIO INICIAL\n")
    mostrar_estado(accion_rojo, accion_azul, puntos_rojo, puntos_azul)

    print("Ambos equipos están decidiendo qué hacer...")
    time.sleep(4)

    # Aplicar reglas
    if accion_rojo == "recoger" and accion_azul == "recoger":
        puntos_rojo += 50
        puntos_azul += 50
        resultado = "Ambos equipos recogieron recursos y ambos obtuvieron puntos."

    elif accion_rojo == "recoger" and accion_azul == "atacar":
        puntos_rojo += 10
        puntos_azul += 70
        resultado = "El equipo azul atacó mientras el rojo recogía. Azul obtuvo ventaja."

    elif accion_rojo == "atacar" and accion_azul == "recoger":
        puntos_rojo += 70
        puntos_azul += 10
        resultado = "El equipo rojo atacó mientras el azul recogía. Rojo obtuvo ventaja."

    else:
        puntos_rojo += 20
        puntos_azul += 20
        resultado = "Ambos equipos atacaron. Hubo conflicto y ambos ganaron poco."

    clear_output(wait=True)
    print("TEORÍA DE JUEGOS - COMPETENCIA POR RECURSOS\n")
    print("PASO 2: RESULTADO DEL ESCENARIO\n")
    mostrar_estado(accion_rojo, accion_azul, puntos_rojo, puntos_azul)

    print("Resultado:")
    print(resultado)

    time.sleep(6)

clear_output(wait=True)
print("TODOS LOS ESCENARIOS FUERON EVALUADOS\n")
print("Conclusión:")
print("El resultado cambia según la estrategia elegida por cada equipo.")
print("Eso demuestra la idea principal de la teoría de juegos.")
```
