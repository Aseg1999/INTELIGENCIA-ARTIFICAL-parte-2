```python
import time
from IPython.display import clear_output

# ======================================
# TEORÍA DE JUEGOS - TODAS LAS OPCIONES
# ======================================

escenarios = [
("atacar","atacar"),
("atacar","defender"),
("defender","atacar"),
("defender","defender")
]

def barra_vida(vida):
    bloques = vida // 10
    return "🟩"*bloques + "⬜"*(10-bloques)

def mostrar(j1,j2,vida1,vida2):

    print("╔══════════════════════════════════╗")
    print("║      COMBATE ESTRATÉGICO         ║")
    print("╚══════════════════════════════════╝\n")

    print("Jugador 1 ⚔️")
    print("Estrategia:",j1)
    print("Vida:",vida1,barra_vida(vida1))
    print()

    print("Jugador 2 🛡️")
    print("Estrategia:",j2)
    print("Vida:",vida2,barra_vida(vida2))
    print()

for j1,j2 in escenarios:

    vida1 = 100
    vida2 = 100

    clear_output(wait=True)

    print("EVALUANDO ESTRATEGIAS\n")

    mostrar(j1,j2,vida1,vida2)

    time.sleep(3)

    if j1=="atacar" and j2=="atacar":
        vida1 -= 30
        vida2 -= 30
        resultado = "Ambos atacaron, ambos reciben daño."

    elif j1=="atacar" and j2=="defender":
        vida2 -= 10
        resultado = "Jugador 1 ataca, jugador 2 se defiende."

    elif j1=="defender" and j2=="atacar":
        vida1 -= 10
        resultado = "Jugador 2 ataca, jugador 1 se defiende."

    else:
        resultado = "Ambos se defienden, no pasa nada."

    time.sleep(2)

    clear_output(wait=True)

    print("RESULTADO DEL ESCENARIO\n")

    mostrar(j1,j2,vida1,vida2)

    print("Resultado:",resultado)

    time.sleep(5)

print("\nTodos los escenarios fueron evaluados.")
```
