```python
import time
from IPython.display import clear_output

# ==========================================
# HEURÍSTICA VISUAL - EJEMPLO 1
# JUGADOR ELIGE EL TESORO MÁS CERCANO
# ==========================================

# Tamaño del mapa
filas = 7
columnas = 9

# Posiciones
jugador = (3, 1)
tesoro_1 = (1, 7)
tesoro_2 = (5, 4)

# ------------------------------------------
# FUNCIÓN HEURÍSTICA (DISTANCIA MANHATTAN)
# ------------------------------------------
def heuristica(origen, destino):
    return abs(origen[0] - destino[0]) + abs(origen[1] - destino[1])

# ------------------------------------------
# DIBUJAR MAPA
# 😀 = jugador
# 💎 = tesoro
# 🏆 = tesoro elegido
# ⬜ = casilla vacía
# ------------------------------------------
def dibujar_mapa(jugador, tesoro_1, tesoro_2, elegido=None):
    for i in range(filas):
        fila_visual = ""
        for j in range(columnas):
            posicion = (i, j)

            if posicion == jugador:
                fila_visual += "😀 "
            elif posicion == tesoro_1 and elegido == tesoro_1:
                fila_visual += "🏆 "
            elif posicion == tesoro_2 and elegido == tesoro_2:
                fila_visual += "🏆 "
            elif posicion == tesoro_1:
                fila_visual += "💎 "
            elif posicion == tesoro_2:
                fila_visual += "💎 "
            else:
                fila_visual += "⬜ "
        print(fila_visual)
    print()

# ------------------------------------------
# MOSTRAR TÍTULO
# ------------------------------------------
clear_output(wait=True)
print("HEURÍSTICA VISUAL - VIDEOJUEGO")
print("Ejemplo 1: elegir el tesoro más cercano\n")
time.sleep(2)

# ------------------------------------------
# MOSTRAR MAPA INICIAL
# ------------------------------------------
clear_output(wait=True)
print("PASO 1: MAPA INICIAL\n")
dibujar_mapa(jugador, tesoro_1, tesoro_2)

print("Leyenda:")
print("😀 = Jugador")
print("💎 = Tesoro")
print("🏆 = Tesoro elegido")
print("⬜ = Espacio libre")
time.sleep(4)

# ------------------------------------------
# MOSTRAR POSICIONES
# ------------------------------------------
clear_output(wait=True)
print("PASO 2: POSICIONES EN EL MAPA\n")
dibujar_mapa(jugador, tesoro_1, tesoro_2)

print(f"Posición del jugador: {jugador}")
print(f"Posición del tesoro 1: {tesoro_1}")
print(f"Posición del tesoro 2: {tesoro_2}")
time.sleep(4)

# ------------------------------------------
# EXPLICAR CÁLCULO DEL TESORO 1
# ------------------------------------------
d1_filas = abs(jugador[0] - tesoro_1[0])
d1_columnas = abs(jugador[1] - tesoro_1[1])
distancia_1 = heuristica(jugador, tesoro_1)

clear_output(wait=True)
print("PASO 3: CÁLCULO HEURÍSTICO DEL TESORO 1\n")
dibujar_mapa(jugador, tesoro_1, tesoro_2)

print("Fórmula: |fila jugador - fila tesoro| + |columna jugador - columna tesoro|")
print()
print(f"|{jugador[0]} - {tesoro_1[0]}| = {d1_filas}")
print(f"|{jugador[1]} - {tesoro_1[1]}| = {d1_columnas}")
print(f"Distancia heurística al tesoro 1 = {d1_filas} + {d1_columnas} = {distancia_1}")
time.sleep(6)

# ------------------------------------------
# EXPLICAR CÁLCULO DEL TESORO 2
# ------------------------------------------
d2_filas = abs(jugador[0] - tesoro_2[0])
d2_columnas = abs(jugador[1] - tesoro_2[1])
distancia_2 = heuristica(jugador, tesoro_2)

clear_output(wait=True)
print("PASO 4: CÁLCULO HEURÍSTICO DEL TESORO 2\n")
dibujar_mapa(jugador, tesoro_1, tesoro_2)

print("Fórmula: |fila jugador - fila tesoro| + |columna jugador - columna tesoro|")
print()
print(f"|{jugador[0]} - {tesoro_2[0]}| = {d2_filas}")
print(f"|{jugador[1]} - {tesoro_2[1]}| = {d2_columnas}")
print(f"Distancia heurística al tesoro 2 = {d2_filas} + {d2_columnas} = {distancia_2}")
time.sleep(6)

# ------------------------------------------
# COMPARAR RESULTADOS
# ------------------------------------------
clear_output(wait=True)
print("PASO 5: COMPARACIÓN DE DISTANCIAS\n")
dibujar_mapa(jugador, tesoro_1, tesoro_2)

print(f"Distancia al tesoro 1: {distancia_1}")
print(f"Distancia al tesoro 2: {distancia_2}")
print()

if distancia_1 < distancia_2:
    print("La heurística indica que el tesoro 1 está más cerca.")
    elegido = tesoro_1
elif distancia_2 < distancia_1:
    print("La heurística indica que el tesoro 2 está más cerca.")
    elegido = tesoro_2
else:
    print("La heurística indica que ambos tesoros están a la misma distancia.")
    elegido = None

time.sleep(5)

# ------------------------------------------
# RESULTADO FINAL
# ------------------------------------------
clear_output(wait=True)
print("PASO 6: DECISIÓN FINAL DE LA HEURÍSTICA\n")
dibujar_mapa(jugador, tesoro_1, tesoro_2, elegido)

print(f"Distancia final al tesoro 1: {distancia_1}")
print(f"Distancia final al tesoro 2: {distancia_2}")
print()

if elegido == tesoro_1:
    print("Resultado: el jugador debería buscar primero el tesoro 1.")
elif elegido == tesoro_2:
    print("Resultado: el jugador debería buscar primero el tesoro 2.")
else:
    print("Resultado: cualquiera de los dos tesoros puede elegirse.")
```
